## Category And Sub-Category

```python
x=df.groupby(["Category","Sub-Category"]).Sales.sum()
y=df.groupby(["Category","Sub-Category"]).Profit.sum()
```

### Sales

```python
fig=plt.figure()
ax0=fig.add_subplot(2,2,1)
ax1=fig.add_subplot(2,2,2)
ax2=fig.add_subplot(2,2,3)
ax3=fig.add_subplot(2,2,4)
x["Furniture"].sort_values(ascending=True).plot(kind="barh",figsize=(20,20),ax=ax1,color="blue")
ax1.set_title("Furniture")
ax1.set_xlabel("Sub-Category")
ax1.set_ylabel("Sales")
x["Technology"].sort_values(ascending=True).plot(kind="barh",figsize=(20,20),ax=ax2,color="red")
ax2.set_title("Technology")
ax2.set_xlabel("Sub-Category")
ax2.set_ylabel("Sales")
x["Office Supplies"].sort_values(ascending=True).plot(kind="barh",figsize=(20,20),ax=ax3,color="yellow")
ax3.set_title("Office Supplies")
ax3.set_xlabel("Sub-Category")
ax3.set_ylabel("Sales")
x.sort_values(ascending=True).plot(kind="barh",figsize=(20,20),ax=ax0,color="black")
ax0.set_title("All sub categories")
ax0.set_xlabel("Sub-Category")
ax0.set_ylabel("Sales")
```

### Profit

```python
fig=plt.figure()
ax0=fig.add_subplot(2,2,1)
ax1=fig.add_subplot(2,2,2)
ax2=fig.add_subplot(2,2,3)
ax3=fig.add_subplot(2,2,4)
y["Furniture"].sort_values(ascending=True).plot(kind="barh",figsize=(20,20),ax=ax1,color="black")
ax1.set_title("Furniture")
ax1.set_xlabel("Sub-Category")
ax1.set_ylabel("Profit")
y["Technology"].sort_values(ascending=True).plot(kind="barh",figsize=(20,20),ax=ax2,color="blue")
ax2.set_title("Technology")
ax2.set_xlabel("Sub-Category")
ax2.set_ylabel("Profit")
y["Office Supplies"].sort_values(ascending=True).plot(kind="barh",figsize=(20,20),ax=ax3,color="blue")
ax3.set_title("Office Supplies")
ax3.set_xlabel("Sub-Category")
ax3.set_ylabel("Profit")
y.sort_values(ascending=True).plot(kind="barh",figsize=(20,20),ax=ax0,color="black")
ax0.set_title("All sub categories")
ax0.set_xlabel("Sub-Category")
ax0.set_ylabel("Profit")
```

### Combine the Sales And Profit
